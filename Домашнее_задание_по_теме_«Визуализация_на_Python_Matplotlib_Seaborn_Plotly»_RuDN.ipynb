```python
# Добавить свой код в ячейку
import pandas as pd
import os

file_name = 'car_data.csv'
df = pd.read_csv(file_name)
df.head()
```

Для числовых признаков постройте гистограммы с использованием Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
# Добавить свой код в ячейку
import matplotlib.pyplot as plt
numeric_columns =df.select_dtypes(include=['number']).columns
numeric_columns=numeric_columns[1:3]
numeric_columns
for column in numeric_columns:
    plt.figure()
    df[column].hist()
    plt.title(f'Гистограмма для {column}')
    plt.xlabel(column)
    plt.ylabel('Количество')
    plt.show()
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1,2,1)
plt.hist(df['Annual Income'])
plt.title(f'Гистограмма для Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Количество')
plt.subplot(1,2,2)
plt.hist(df['Price ($)'])
plt.title(f'Гистограмма для Price ($)')
plt.hist(df['Price ($)'])
plt.ylabel('Количество')
plt.show()
```

Постройте график типа box plot с горизонтальным размещением ящиков для визуализации зависимости стоимости автомобилей (Price ($)) от марки (Company)  с использованием Matplotlib

```python
# Добавить свой код в ячейку

grouped_data = df.groupby('Company')['Price ($)'].apply(list)

companies = grouped_data.index.tolist()
prices = grouped_data.values.tolist()

plt.figure(figsize=(10,8))
plt.boxplot(prices, vert=False, labels=companies) 
plt.title('Зависимость стоимости автомобилей от марки')
plt.xlabel('Price ($)')
plt.ylabel('Company')

plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.scatter(df, x='Annual Income', y='Price ($)', color='Transmission',
                 title='Зависимость стоимости автомобилей от годового дохода',
                 labels={'Price': 'Price ($)', 'Annual Income': 'Годовой доход'})

# Отображаем график
fig.show()
```

```python
avg_price = df.groupby('Company')['Price ($)'].mean().reset_index()
fig = px.bar(avg_price, x='Company', y='Price ($)', title='Средняя цена автомобиля по маркам')
fig.show()
```
